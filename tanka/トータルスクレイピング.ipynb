```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# メールアドレスとパスワードの指定
USER = "chiha68@gmail.com"
PASS = "yukitan0619"

# セッションを開始
session = requests.session()

# ログイン
login_info = {
    "email":USER,
    "password":PASS,
}

# action
url_login = "https://www.utayom.in/login"
res = session.post(url_login, data=login_info)
res.raise_for_status() # エラーならここで例外を発生させる

print(res.text)
```
